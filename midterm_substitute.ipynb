```python
y_data=np.array(y_data, dtype=np.int32) #float를 one_hot하면 오류가 나므로  int로 전환 
print(y_data)
```

```python
#make y_data one_hot_dummy
one_hot_dummy=tf.one_hot(y_data, depth=3)
```

```python
#int to float
x_data=np.array(x_data, dtype=np.float32)
y_data=np.array(one_hot_dummy, dtype=np.float32)
y_data=np.squeeze(y_data)
print(x_data.shape)
print(y_data.shape)
print(y_data)
print(x_data)

def MinMaxScaler(data):
    numerator=data-np.min(data, 0)
    denominator=np.max(data, 0)-np.min(data, 0) + 1e-7
    return numerator / denominator

x_data=MinMaxScaler(x_data)
print(x_data)

x_train, x_test, y_train, y_test=train_test_split(x_data, y_data, test_size=0.3, random_state=2020)
```

```python
tf.random.set_seed(2000)

W=tf.Variable(tf.random.normal([6, 3], mean=0.0))
b=tf.Variable(tf.random.normal([3], mean=0.0))


def softmax(X):
    z=tf.matmul(X,W)+b
    return tf.nn.softmax(z)
```

```python
def cost_fn(y_pred, y_actual):
    return tf.reduce_mean(-tf.reduce_sum(y_actual*tf.math.log(y_pred), axis=1))

optimizer=tf.optimizers.SGD(learning_rate=0.001)
```

```python
for step in range(500001):  #50만번 진행
    with tf.GradientTape() as g:
        pred=softmax(x_train)
        cost=cost_fn(pred, y_train)
        
        gradients=g.gradient(cost, [W,b])
        
    optimizer.apply_gradients(zip(gradients, [W,b]))
    
    if step % 100000 == 0:
        print("cost: \t", cost.numpy())
        print("W: \t", W.numpy())
        print("b: \t", b.numpy())
        print("\n")
```

```python
# train accuracy
predicted=tf.argmax(softmax(x_train), axis=1)
real=tf.argmax(y_train, axis=1)

def accuracy(y_predicted, y_real):
    return tf.reduce_mean(tf.cast(tf.equal(y_predicted, y_real), dtype = tf.float32))

print("train accuracy: ", accuracy(predicted, real).numpy())
```

```python
# test accuracy
tested=tf.argmax(softmax(x_test), axis=1)
reall=tf.argmax(y_test, axis=1)
print("test accuracy: ", accuracy(tested, reall).numpy())
```
