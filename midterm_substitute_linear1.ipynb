```python
import tensorflow as tf
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
df=pd.read_csv(r"C:\Juneinter\midterm_substitution.csv")
df
```

```python
infla_x_data=df.iloc[:,[3]].values
debt_x_data=df.iloc[:,[4]].values
growth_y_data=df.iloc[:,[2]].values
```

```python
W=tf.Variable(tf.random.normal([1], mean=0.0), name="weight")
b=tf.Variable(tf.random.normal([1], mean=0.0), name="bias")

def hypothesis(X):
    return W*X+b

def cost_fn(y_pred, y_true):
    return tf.reduce_mean(tf.square(y_pred-y_true))

optimizer=tf.optimizers.SGD(learning_rate=0.001)
```

```python
for step in range(10001):  #인플레이션과 경제성장률 상관분석
    with tf.GradientTape() as g:
        pred=hypothesis(infla_x_data)
        cost=cost_fn(pred, growth_y_data)
        
        gradients=g.gradient(cost, [W,b])
        
    optimizer.apply_gradients(zip(gradients, [W,b]))
    
    if step % 2500==0:
        print("cost: ", cost.numpy())
        print("slope: ", W.numpy())
        print("y-intercept: ", b.numpy())
```
