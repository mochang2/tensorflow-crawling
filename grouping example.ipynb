```python
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
#평균 (1, 1)이고 분산이 단위행렬인 정규분포에서 데이터 (100,2) 추출
G0 = np.random.multivariate_normal([1, 1], np.eye(2), 100) # (mean, cov, size), (100,2)
#평균 (3, 5)이고 분산이 단위행렬인 정규분포에서 데이터 (100,2) 추출
G1 = np.random.multivariate_normal([3, 5], np.eye(2), 100)
#평균 (9, 9)이고 분산이 단위행렬인 정규분포에서 데이터 (100,2) 추출
G2 = np.random.multivariate_normal([9, 9], np.eye(2), 100)
```

```python
X = np.vstack([G0,G1,G2])
X = np.asmatrix(X)
print(G0.shape, X.shape) # (100,2) (300,2)
```

```python
plt.figure(figsize=(12,5))   #size 12, 5 inches
plt.subplot(1,2,1)           #1개 행 2개 열의 1번째
plt.plot(X)
plt.subplot(1,2,2)           #1개 행 2개 열의 2번째
plt.plot(X[:,0], X[:,1], 'b.')
plt.show()
```

```python
plt.figure(figsize = (10, 8))
plt.plot(G0[:,0], G0[:,1], 'b.', label = 'G0')
plt.plot(G1[:,0], G1[:,1], 'g.', label = 'G1')
plt.plot(G2[:,0], G2[:,1], 'r.', label = 'G2')
plt.axis('equal')
plt.legend(fontsize = 12)
plt.show()
```
